```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
url = '/content/insurance.csv'
data= pd.read_csv(url)
data.sample(5)
data.info()
print(data['sex'].value_counts())
sns.countplot(x='sex',data=data)
print(data['smoker'].value_counts())
sns.countplot(x='smoker',data=data)
print(data['region'].value_counts())
sns.countplot(x='region',data=data)
corr=data.corr()
sns.heatmap(corr,cmap='Wistia',annot=True)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['age']),y=data['charges'])
print('He so Intercept cua bac 1 la : ',model.intercept_)
print ('He so coeffient  cua bac 1 la : ', model.coef_)
model.score(X=pd.DataFrame(data['age']),y=data['charges'])
y_pred= model.predict(X=pd.DataFrame(data['age']))
data.plot(kind='scatter',x='age',
 y='charges',figsize=(9,9),
 color='black',xlim=(0,400))
```

```python
model = linear_model. LinearRegression()
X=pd.DataFrame([data['age'],data['age']**2,data['age']**3,data['age']**4,data['age']**5]).T
model.fit ( X ,y=data['charges'])
print ('Intercept cua mo hinh voi bac 5 :',model.intercept_)
print('Coeff cua mo dinh voi bac 5 :',model.coef_)
print('Do chinh sat cua mo hinh voi bac 5 :',model.score(X,y=data['charges']))

range = np.arange (0,300 ,0.1)
X_range = pd.DataFrame ([range,range**2,range**3,range**4,range**5]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='age',
          y='charges',figsize=(9,9),
          color='black',xlim=(0,400))
```
